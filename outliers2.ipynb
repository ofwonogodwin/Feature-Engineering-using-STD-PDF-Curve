Outlier Detection And Removal Using Standard Deviation

```python
#Using Standard Deviation (3 std)
#In this Method we use the concept of normal Distribution Learnt In Probability Class to detect the outliers
#Nice Fact, My Teacher Of Probability In The University Is Mr.Simon Kabito
#We draw a normal Distribution Curve That Enables us To Visualize Which dataPoints are Outliers
```

```python
#We First Import The needed libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('height2.csv')
df.sample(5)
```

```python
#Checking The Number Of Rows and Columns Respectively
df.shape
```

#Plotting the Histogram

```python
plt.hist(df.Height, bins = 20,rwidth = 0.8, color ='green')
plt.xlabel('Height In Inches')
plt.ylabel('Number Of People With The Particular Heights')
plt.title('Simple Histogram To Visualize')
plt.show
#Fun fact, I love Green for Visualizing Data
```

```python
#Here We are trying To import what Is necessary For Us to Plot The pdf Curve
from scipy.stats import norm
import numpy as np
```

```python
#PDF  >> Probability Density Function
x_range = np.linspace(df.Height.min(), df.Height.max(), 1000)  
# Using 1000 evenly spaced values
# We now compute the PDF based on mean and standard deviation of Height Column
pdf_values = norm.pdf(x_range, df.Height.mean(), df.Height.std())
```

            #Plotting The Histogram and The Normal Distribution Curve Over It

```python
#Plotting the PDF curve
#The alpha simply determines the Transparency Of our Histogram
plt.hist(df.Height, bins=20, density=True, rwidth=0.8,alpha=0.8, color='green', label='Histogram')
# Plot the PDF
plt.plot(x_range, pdf_values, label='Normal Distribution Curve',color='black')
plt.xlabel('Height')
plt.ylabel('Density')
plt.title('Probability Density Function')
plt.legend(loc='best')
plt.show()
```

```python

#Checking The mean of The Height Column
df.Height.mean()
```

```python
#Checking The standard deviation of The Height Column
df.Height.std()
```

```python
#Kindly Google why We use 3 STD and How it actually Operates
#Using 3 std to remove Outliers. U can also use 4 std if you want
#We are Simply Going to add The Mean with three times the STD
upper_limit = df.Height.mean() + 3*df.Height.std()
upper_limit
```

```python
#We are Simply Going to subtract three times the STD from Mean
lower_limit = df.Height.mean() - 3*df.Height.std()
lower_limit
```

```python
#Now we are Displaying The Outliers
df[(df.Height>upper_limit) | (df.Height<lower_limit)]
```

```python
#Since we know the outliers, We can now Display Our Clean Dataset
# We are going to name our new Dataset df2
df2 = df[(df.Height<upper_limit)&(df.Height>lower_limit)]
df2.head()
```
